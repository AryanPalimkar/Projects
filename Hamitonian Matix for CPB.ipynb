Function for generating the tunneling Hamiltonian matrix(Coulomb energy included):

```python
import numpy as np
import matplotlib.pyplot as plt


def Hamiltonian(size, E_j,E_c, n_g, V):
    H=np.zeros((size,size),dtype=float)
    if(size%2==0):
        try:
            raise ValueError()
        except ValueError:
            print("The size argument must be an odd number")
    
   
    for n in range (size):
        for m in range (size):
            H[n,m]=-E_j/2 * 1 if (n == (m-1) or n == (m+1)) else 0
            if n==m : H[n,m]=4*E_c*((m-int(size/2))**2-2*(m-int(size/2))*n_g+1)
            
    print(H)
        
    return H
            
            
            
            
        
```

```python
s=21
E_j=80 #Input the value of Josephson coupling energy
E_c=1 #Input the value of Coulombic energy associated with transfer of one electron across the junction
n_g=0.5 #Input the residual charge of the electrode
V=1 #Inpt potential difference across junctions
H=Hamiltonian(s,E_j,E_c,n_g,V)
```

```python
w,v=np.linalg.eig(H)
# print(v)
# print(v[:,0])
# print(w[0])
# print(w[0]*v[:,0])
# print(np.matmul(H,v[:,0]))
# if((np.matmul(H,v[:,0])==w[0]*v[:,0]).all()):
#     print("True..works")
# else:
#     print("Check again")
```

```python
def plotting_eigenvalues(w,v,i,size):
    #plt.figsize((10,6))
    x=[i for i in range(size)]
    x=[a-int(size/2) for a in x]
    y=v[:,i]
    plt.bar(x,y)
    plt.title(f"Plot of an eigenvector associated with the eigenvalue {w[i]} in number basis")
    plt.xlabel('component in number basis')
    plt.ylabel('strength of the component in the eigenvector')
    plt.grid(color='black',linestyle='-',linewidth='1')
    plt.show()
    
    
```

```python
for i in range(s):
    plotting_eigenvalues(w,v,i,s)
```
