Function for generating the tunneling Hamiltonian matrix(Coulomb energy included):

```python
import numpy as np
import cmath
import math
import matplotlib.pyplot as plt


def Hamiltonian(size, E_j,E_c,phi_e,gamma): 
    H=np.zeros((size,size),dtype=complex)
    if(size%2==0):
        try:
            raise ValueError()
        except ValueError:
            print("The size argument must be an odd number")
    
   
    for n in range (size):
        for m in range (size):
            if n==m:
                H[n,m]=4*E_c*(m**2)
            elif n==m+1:
                H[n,m]=-E_j/2*(np.cos(phi_e)-1j*np.sin(phi_e)+2*gamma)
            elif n==m-1:
                H[n,m]=-E_j/2*(np.cos(phi_e)+1j*np.cos(phi_e)+2*gamma)
            
    print(H)
        
    return H     
        
```

```python
s=11

E_j=80     #Input the value of Josephson coupling energy

E_c=1      #Input the value of Coulombic energy associated with transfer of one electron across the junction

phi_e=np.pi/2 #The external flux used to tune the effective Josephson energy

gamma=2    #Strength of the array junctions relative to the single junctions

H=Hamiltonian(s,E_j,E_c,phi_e,gamma)
```

```python
def plotting_eigenvalues(w,v,i,size):
    #plt.figsize((10,6))
    x=[i for i in range(size)]
    x=[a-int(size/2) for a in x]
    y=v[:,i]
    plt.bar(x,y)
    plt.title(f"Plot of an eigenvector associated with the eigenvalue {w[i]} in number basis")
    plt.xlabel('component in number basis')
    plt.ylabel('strength of the component in the eigenvector')
    plt.grid(color='black',linestyle='-',linewidth='1')
    plt.show()
    
    
```

```python
for i in range(s):
    plotting_eigenvalues(w,v,i,s)
```
